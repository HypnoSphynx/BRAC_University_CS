```python
#generate large prime(128 bits) p and q
import random
from sympy import isprime

def generate_prime(bits):
    while True:
        num = random.getrandbits(bits)
        num |= (1 << (bits - 1)) | 1
        if isprime(num):
            return num

# Generate two 128-bit prime numbers
p = generate_prime(128)
q = generate_prime(128)

print(f"p: {p}")
print(f"q: {q}")
```

```python
#Calculate n = pq

n=p*q
print(n)
```

```python
#Calculate phi(n)=(p-1)(q-1)
phi_n=(p-1)*(q-1)
print(phi_n)
```

```python
#Perform encryption here and generate ciphertext [c = m^e (mod n)] and private key [d*e â‰¡ 1 (mod(phi(n)))]
e = 11
d= pow(e,-1,phi_n)
print(d)

c=pow(m,e,n)
print(c)
```

```python
#perform decryption here [m = c^d (mod n)]

m= pow(c,d,n)
print(m)
```
