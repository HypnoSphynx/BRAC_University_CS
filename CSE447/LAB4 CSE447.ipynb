```python
#1
import os
import sys
import binascii

from cryptography.hazmat.primitives import cmac
from cryptography.hazmat.decrepit.ciphers import algorithms

message = "I met an interesting turtle while the song on the radio blasted away"
message_bytes = message.encode()  # Convert string to bytes
key = b'\x01\xd8i\xa1^0\x9a<\x0f\xf0\r\xc1\xdd\xd5\x89\xa6'
# You code here. HINT: use CMAC, TripleDES
mac= cmac.CMAC(algorithms.TripleDES(key))
get_signature=mac.update(message_bytes)
signature=mac.finalize()

print(f"Message: {message}")
print(f"CMAC signature: {binascii.b2a_hex(signature).decode()}")

given_signature = "ba4ecb8db45c6ae0"


if signature.hex() == given_signature:
    print("Valid")
else:
    print("Invalid")
```

```python
import operator
from statistics import mode

def majority(bits):
    return mode(bits)

def a51(x, y, z, n):
    key = ''
    x = [int(bit) for bit in x]
    y = [int(bit) for bit in y]
    z = [int(bit) for bit in z]

    # Length checks
    if len(x) != 19:
        print("Input of X is invalid")
        return
    if len(y) != 22:
        print("Input of Y is invalid")
        return
    if len(z) != 23:
        print("Input of Z is invalid")
        return

    for i in range(n):

        majority_bit = majority([x[8], y[10], z[10]])


        if x[8] == majority_bit:
            x_input = x[13] ^ x[16] ^ x[17] ^ x[18]
            x = [x_input] + x[:-1]


        if y[10] == majority_bit:
            y_input = y[20] ^ y[21]
            y = [y_input] + y[:-1]


        if z[10] == majority_bit:
            z_input = z[7] ^ z[20] ^ z[21] ^ z[22]
            z = [z_input] + z[:-1]

        key_bit = x[18] ^ y[21] ^ z[22]
        key += str(key_bit)


    return key
```

```python
def encrypt(plaintext, keystream):
    binary = ''.join(format(ord(c), '08b') for c in plaintext)
    return ''.join(str(int(b) ^ int(k)) for b, k in zip(binary, keystream))

def decrypt(ciphertext, keystream):
    binary = ''.join(str(int(c) ^ int(k)) for c, k in zip(ciphertext, keystream))
    return ''.join(chr(int(binary[i:i+8], 2)) for i in range(0, len(binary), 8))
```

```python
strings = [
    'It is alive','Snap out of it','I am as mad as hell and I am not going to take this anymore','Bond James Bond',"Love means never having to say you're sorry"
]

x = '1110001100101001011'
y = '0011000000010000001101'
z = '10011101101111001001110'

for i, plain in enumerate(strings):
    binary_len = len(plain) * 8
    key_stream = a51(x, y, z, binary_len)

    ciphertext = encrypt(plain, key_stream)
    recovered = decrypt(ciphertext, key_stream)

    print(f"Plaintext : {plain}")
    print(f"Ciphertext: {ciphertext}")
    print(f"Recovered : {recovered}")
    print("-"*60)
```
