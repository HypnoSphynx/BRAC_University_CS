```python
class Stack():
    def __init__(self,len):
        self.stack=[None]*len
        self.size=0
    def is_empty(self):
        if self.size==0:
            return True
    def is_full(self):
        if self.size>=len(self.stack):
            return True  
    def push(self,elem):
        if self.is_full()==True:
            return 'Overflow Exception'
        self.stack[self.size]=elem
        self.size+=1
    def peek(self):
        if self.is_empty()==True:
            return 'Underflow exception'
        return self.stack[self.size-1]
    def pop(self):
        if self.is_empty()==True:
            return 'Underflow exception'
        temp=self.stack[self.size-1]
        self.stack[self.size-1]=None
        self.size-=1
        return temp
```

```python
#Stack with linked list
class Node:
    def __init__(self,e,n,p):
        self.elem=e
        self.next=n
        self.prev=p
class Linkedlist:
    def __init__(self):
            self.head=Node(None,None,None)

class stack:
    def __init__(self):
        self.lst=Linkedlist()
        self.node_at=self.lst.head
        self.count=0
    
    def push(self,value):
        self.node_at.next=Node(value,None,self.node_at)
        self.node_at=self.node_at.next
        self.count+=1
    def top(self):
        if self.count!=0:
            temp=self.node_at.elem
            self.node_at=self.node_at.prev
            self.node_at.next.prev=None
            self.node_at.next=None
            return temp

        else:
            print('Underflow exception')
    def peek(self):
        if self.count!=0:
            return self.node_at.elem
        else:
            print('Underflow exception')
```
