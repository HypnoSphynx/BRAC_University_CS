```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python

class DoublyList:
  
    def __init__(self, a):
      if type(a)==list:

          self.head=Node(a[0],None,None)
          tail=self.head

          for i in range(1,len(a)):
              new_node=Node(a[i],None,tail)
              tail.next=new_node
              tail=new_node
            
          self.head.prev=tail
          tail.next=self.head
      else:
          self.head=a

    
  # Counts the number of Nodes in the list
    def countNode(self):
        temp=self.head.next
        count=1
        while temp!=self.head:
            count+=1
            temp=temp.next
        return count
        


  
  # prints the elements in the list
    def forwardprint(self):
        temp=self.head.next
        output= f'{self.head.element}'

        while temp!=self.head:
            output+=' '+str(temp.element)
            temp=temp.next
        print(output)
        


  # prints the elements in the list backward
    def backwardprint(self):
        temp=self.head.prev
        output=''
        while temp !=self.head:
            if temp==self.head.prev:
                output+=str(temp.element)
            else:
                output+=' '+str(temp.element)
            temp=temp.prev
        output+=' '+str(self.head.element)
        print(output)




    # returns the reference of the at the given index. For invalid index return None.
    def nodeAt(self, idx):
        temp=self.head.next
        count=1
        if 0<idx>self.countNode():
            return Node(None,None,None)
        if idx==0:
            return self.head
        while temp!=self.head:
            if count==idx:
                return temp
            count+=1
            temp=temp.next
    
    # returns the index of the containing the given element. if the element does not exist in the List, return -1.
    def indexOf(self, elem):
        temp=self.head.next
        count=1
        if elem==self.head.element:
            return 0
        while temp!=self.head:
            if elem==temp.element:
                return count
            count+=1
            temp=temp.next
        
        return -1
            
  

  # inserts containing the given element at the given index Check validity of index. 
    def insert(self, elem, idx):
         temp=self.head
         index=0
         if 0<idx>self.countNode():
            print('Invalid Index')
        #Using For loop.
        #  for i in range(idx+1):
        #     if index==idx:
        #         temp=temp.prev
        #         node=Node(elem,temp.next,temp)
        #         temp.next.prev=node
        #         temp.next=node
        #     index+=1
        #     temp=temp.next

        #using while loop.
         while index<=idx:
            if index==idx:
                temp=temp.prev
                node=Node(elem,temp.next,temp)
                temp.next.prev=node
                temp.next=node
                if idx==0:
                    self.head=temp.next
                break
            index+=1
            temp=temp.next



  # removes at the given index. returns element of the removed node. Check validity of index. return None if index is invalid.
    def remove(self, idx):
        temp=self.head
        index=0
        output=None
        
        if 0 > idx > (self.countNode()-1):
            return None
        while index<=idx:
            if index==idx:
                output=temp.element
                temp=temp.prev
                temp.next=temp.next.next
                temp.next.prev=temp
                if idx==0:
                    self.head=temp.next
                return str(output)
            index+=1
            temp=temp.next
        
 
    
```
